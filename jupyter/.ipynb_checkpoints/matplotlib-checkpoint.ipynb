```python
import math
import numpy as np
import matplotlib.pyplot as plt
pi = math.pi   #mathモジュールのπを利用

x = np.linspace(0, 2*pi, 100)  #0から2πまでの範囲を100分割したnumpy配列
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
# 三角関数で距離を求める
naname = 50
kakudo = 30
cos = math.cos(math.radians(kakudo))
sin = math.sin(math.radians(kakudo))
print(f'対象物との距離は、{naname*cos}です。')
print(f'対象物の高さは、{naname*sin}です。')
```
